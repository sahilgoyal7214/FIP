Write a program to blur an image using Gaussian low pass filter and then sharpen the
image using sharpening filter.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def apply_gaussian_blur(image, kernel_size):
    return cv2.GaussianBlur(image, (kernel_size, kernel_size), 0)

def sharpen_image(image, alpha, beta):
    blurred = cv2.GaussianBlur(image, (0, 0), 3)
    sharpened = cv2.addWeighted(image, alpha, blurred, beta, 0)
    return sharpened

# Read the original image
image_path = r'C:\Users\Suyash Tambe\Desktop\FIPL ESE\20230418_102332.jpg'
original_image = cv2.imread(image_path)

# Apply Gaussian blur
blurred_image = apply_gaussian_blur(original_image, kernel_size=5)

# Sharpen the blurred image
sharpened_image = sharpen_image(blurred_image, alpha=1.5, beta=-0.5)

# Plot the original, blurred, and sharpened images
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB))
plt.title('Blurred Image')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(sharpened_image, cv2.COLOR_BGR2RGB))
plt.title('Sharpened Image')

plt.tight_layout()
plt.show()
```
