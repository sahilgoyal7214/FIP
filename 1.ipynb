1. Reading and displaying RGB image and convert to gray-scale and display. Write
programs using any two libraries.

```python
import cv2
import matplotlib.pyplot as plt

# Read RGB image
rgb_image = cv2.imread(r'C:\Users\Suyash Tambe\Desktop\FIPL ESE\20230418_102332.jpg')

# Display RGB image
plt.imshow(cv2.cvtColor(rgb_image, cv2.COLOR_BGR2RGB))
plt.title('RGB Image')
plt.axis('off')
plt.show()

# Convert RGB to grayscale
gray_image = cv2.cvtColor(rgb_image, cv2.COLOR_BGR2GRAY)

# Display grayscale image
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```

```python
from PIL import Image
import matplotlib.pyplot as plt

# Read RGB image
rgb_image_path = r'C:\Users\Suyash Tambe\Desktop\FIPL ESE\20230418_102332.jpg'
rgb_image = Image.open(rgb_image_path)

# Display RGB image
plt.imshow(rgb_image)
plt.title('RGB Image')
plt.axis('off')
plt.show()

# Convert RGB to grayscale
gray_image = rgb_image.convert('L')

# Display grayscale image
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```
