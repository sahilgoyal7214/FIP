Write program to obtain edges of an image using Prewitt operator. Plot horizontal,
vertical and combined edge detected images.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def prewitt_operator(image):
    kernel_x = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
    kernel_y = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])

    # Apply convolution with Prewitt operators
    gradient_x = cv2.filter2D(image, cv2.CV_64F, kernel_x)
    gradient_y = cv2.filter2D(image, cv2.CV_64F, kernel_y)

    # Compute magnitude of gradients
    magnitude = np.sqrt(np.square(gradient_x) + np.square(gradient_y))

    return gradient_x, gradient_y, magnitude

# Read the original image
image_path = r'C:\Users\Suyash Tambe\Desktop\FIPL ESE\20230418_102332.jpg'
original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Apply Prewitt operator
gradient_x, gradient_y, magnitude = prewitt_operator(original_image)

# Plot the results
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.imshow(original_image, cmap='gray')
plt.title('Original Image')

plt.subplot(2, 2, 2)
plt.imshow(gradient_x, cmap='gray')
plt.title('Gradient X')

plt.subplot(2, 2, 3)
plt.imshow(gradient_y, cmap='gray')
plt.title('Gradient Y')

plt.subplot(2, 2, 4)
plt.imshow(magnitude, cmap='gray')
plt.title('Combined Edges')

plt.tight_layout()
plt.show()
```
