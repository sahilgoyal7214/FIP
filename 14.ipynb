Write a program to perform opening and closing on an image.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def apply_opening(image, kernel_size):
    kernel = np.ones((kernel_size, kernel_size), np.uint8)
    opened_image = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)
    return opened_image

def apply_closing(image, kernel_size):
    kernel = np.ones((kernel_size, kernel_size), np.uint8)
    closed_image = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)
    return closed_image

# Read the original image
image_path = r'C:\Users\Suyash Tambe\Desktop\FIPL ESE\20230418_102332.jpg'
original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Set the kernel size (adjust as needed)
kernel_size = 5

# Apply opening
opened_image = apply_opening(original_image, kernel_size)

# Apply closing
closed_image = apply_closing(original_image, kernel_size)

# Plot the results
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(original_image, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 3, 2)
plt.imshow(opened_image, cmap='gray')
plt.title('Opened Image')

plt.subplot(1, 3, 3)
plt.imshow(closed_image, cmap='gray')
plt.title('Closed Image')

plt.tight_layout()
plt.show()
```
